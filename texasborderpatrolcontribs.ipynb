```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
```

```python
all_contribs = pd.read_csv("~/documents/ca_elections/2024/2019to2024.tex.cbp.contribs.csv")
```

```python
all_contribs.info()
```

```python
all_contribs.groupby("contributor_employer").contribution_receipt_amount.sum().sort_values(ascending=False).reset_index()
```

```python
## elimate employers that are not border patrol:
## CBPP
## CBPS
## CBPA
## DIMENSION CBP, LLC
## CBPD TCGC
```

```python
## updated to eliminate the above employers

updated_contribs = all_contribs[~(all_contribs.contributor_employer.isin(["CBPP", "CBPS","CBPA","DIMENSION CBP, LLC","CBPD TCGC"]))]
```

```python
updated_contribs.value_counts("contributor_employer").reset_index()
```

```python
## chart 1 -- all contribs from 2019 to 2024
```

```python
all_top_contribs = updated_contribs.groupby("committee_name").contribution_receipt_amount.sum().sort_values(ascending=False).reset_index()
```
